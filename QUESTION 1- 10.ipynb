````python
import random

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

def calculate_weighted_mean(data, weights):
    """
    Calculates the weighted mean of a list of values and their corresponding weights.

    Args:
      data: A list of values.
      weights: A list of corresponding weights for each value in `data`.

    Returns:
      The weighted mean of the data.
    """
    if len(data) != len(weights):
        raise ValueError("Data and weights must have the same length.")

    weighted_sum = sum(value * weight for value, weight in zip(data, weights))
    sum_of_weights = sum(weights)
    return weighted_sum / sum_of_weights

# Example usage:
# Assuming you have a list of weights (replace with actual weights)
weights = [1] * len(int_list)  # Example: Equal weights for all values

weighted_mean = calculate_weighted_mean(int_list, weights)
print("Weighted Mean:", weighted_mean)
````

3  Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.

````python
class DiscreteRandomVariable:
    """
    Represents a discrete random variable with methods to calculate expected value and variance.

    Attributes:
        pmf (dict): Probability mass function, where keys are values and values are probabilities.
    """

    def __init__(self, pmf):
        """
        Initializes the DiscreteRandomVariable object.

        Args:
            pmf (dict): Probability mass function.
        """
        self.pmf = pmf

    def expected_value(self):
        """Calculates the expected value of the random variable."""
        expected_value = 0
        for value, probability in self.pmf.items():
            expected_value += value * probability
        return expected_value

    def variance(self):
        """Calculates the variance of the random variable."""
        mean = self.expected_value()
        variance = 0
        for value, probability in self.pmf.items():
            variance += probability * (value - mean) ** 2
        return variance

# Example usage:
pmf = {1: 0.2, 2: 0.3, 3: 0.5}
rv = DiscreteRandomVariable(pmf)

print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
````

4 Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

````python
import random

def simulate_die_rolls(num_rolls):
    """
    Simulates rolling a fair six-sided die multiple times.

    Args:
        num_rolls: The number of times to roll the die.

    Returns:
        A list of the outcomes of each die roll.
    """
    rolls = []
    for _ in range(num_rolls):
        roll = random.randint(1, 6)
        rolls.append(roll)
    return rolls

def calculate_expected_value(rolls):
    """
    Calculates the expected value of a list of die rolls.

    Args:
        rolls: A list of die roll outcomes.

    Returns:
        The expected value of the die rolls.
    """
    return sum(rolls) / len(rolls)

def calculate_variance(rolls):
    """
    Calculates the variance of a list of die rolls.

    Args:
        rolls: A list of die roll outcomes.

    Returns:
        The variance of the die rolls.
    """
    mean = calculate_expected_value(rolls)
    squared_differences = [(roll - mean) ** 2 for roll in rolls]
    return sum(squared_differences) / len(rolls)

if __name__ == "__main__":
    num_rolls = 10000  # Number of times to roll the die
    rolls = simulate_die_rolls(num_rolls)

    expected_value = calculate_expected_value(rolls)
    variance = calculate_variance(rolls)

    print(f"Expected Value: {expected_value:.2f}")
    print(f"Variance: {variance:.2f}")
````

5  Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_random_samples(distribution, params, sample_size=1000):
  """
  Generates random samples from a given probability distribution.

  Args:
    distribution: The name of the distribution (e.g., 'binomial', 'poisson', 'uniform').
    params: A dictionary containing the parameters of the distribution. 
             For example:
             - binomial: {'n': 10, 'p': 0.5}
             - poisson: {'lambda': 3}
             - uniform: {'low': 0, 'high': 1}
    sample_size: The number of samples to generate.

  Returns:
    A list of random samples.
  """

  if distribution == 'binomial':
    n = params['n']
    p = params['p']
    samples = np.random.binomial(n, p, size=sample_size)
  elif distribution == 'poisson':
    lam = params['lambda']
    samples = np.random.poisson(lam, size=sample_size)
  elif distribution == 'uniform':
    low = params['low']
    high = params['high']
    samples = np.random.uniform(low, high, size=sample_size)
  else:
    raise ValueError("Invalid distribution. Supported distributions: 'binomial', 'poisson', 'uniform'")

  return samples

def calculate_mean_and_variance(samples):
  """
  Calculates the mean and variance of a list of samples.

  Args:
    samples: A list of sample values.

  Returns:
    A tuple containing the mean and variance of the samples.
  """
  mean = np.mean(samples)
  variance = np.var(samples)
  return mean, variance

# Example usage:
# Generate 1000 samples from a binomial distribution with n=10 and p=0.3
binomial_samples = generate_random_samples('binomial', {'n': 10, 'p': 0.3})
binomial_mean, binomial_variance = calculate_mean_and_variance(binomial_samples)
print("Binomial Distribution:")
print("Mean:", binomial_mean)
print("Variance:", binomial_variance)

# Generate 1000 samples from a Poisson distribution with lambda=5
poisson_samples = generate_random_samples('poisson', {'lambda': 5})
poisson_mean, poisson_variance = calculate_mean_and_variance(poisson_samples)
print("\nPoisson Distribution:")
print("Mean:", poisson_mean)
print("Variance:", poisson_variance)

# Generate 1000 samples from a uniform distribution between 0 and 1
uniform_samples = generate_random_samples('uniform', {'low': 0, 'high': 1})
uniform_mean, uniform_variance = calculate_mean_and_variance(uniform_samples)
print("\nUniform Distribution:")
print("Mean:", uniform_mean)
print("Variance:", uniform_variance)
````

6 Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute 
the mean, variance, and standard deviation of the samples.

````python
import numpy as np

def generate_gaussian_samples(mu, sigma, sample_size):
  """
  Generates random samples from a Gaussian (normal) distribution.

  Args:
    mu: Mean of the Gaussian distribution.
    sigma: Standard deviation of the Gaussian distribution.
    sample_size: Number of samples to generate.

  Returns:
    A list of generated samples.
  """
  return np.random.normal(mu, sigma, sample_size)

def calculate_statistics(samples):
  """
  Calculates the mean, variance, and standard deviation of a list of samples.

  Args:
    samples: A list of sample values.

  Returns:
    A tuple containing the mean, variance, and standard deviation.
  """
  mean = np.mean(samples)
  variance = np.var(samples)
  std_dev = np.std(samples)
  return mean, variance, std_dev

if __name__ == "__main__":
  mu = 0  # Mean of the Gaussian distribution
  sigma = 1  # Standard deviation of the Gaussian distribution
  sample_size = 1000  # Number of samples to generate

  samples = generate_gaussian_samples(mu, sigma, sample_size)
  mean, variance, std_dev = calculate_statistics(samples)

  print(f"Mean: {mean:.4f}")
  print(f"Variance: {variance:.4f}")
  print(f"Standard Deviation: {std_dev:.4f}")
````

8) Write a Python function to calculate the probability density function (PDF) of a continuous random 
variable for a given normal distribution.

````python
import math

def normal_pdf(x, mu, sigma):
  """
  Calculates the probability density function (PDF) of a normal distribution.

  Args:
    x: The value for which to calculate the PDF.
    mu: The mean of the normal distribution.
    sigma: The standard deviation of the normal distribution.

  Returns:
    The PDF value at the given x.
  """
  term1 = 1 / (sigma * math.sqrt(2 * math.pi))
  term2 = math.exp(-0.5 * ((x - mu) / sigma) ** 2)
  return term1 * term2

# Example usage:
x = 1.5  # Value for which to calculate the PDF
mu = 0    # Mean of the normal distribution
sigma = 1  # Standard deviation of the normal distribution

pdf_value = normal_pdf(x, mu, sigma)
print(f"PDF at x={x}: {pdf_value:.4f}")
````

9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python
import numpy as np

def exponential_cdf(x, rate):
  """
  Calculates the cumulative distribution function (CDF) of the exponential distribution.

  Args:
    x: The value at which to evaluate the CDF.
    rate: The rate parameter of the exponential distribution (lambda).

  Returns:
    The value of the CDF at x.
  """
  if x < 0:
    return 0.0
  else:
    return 1 - np.exp(-rate * x)

# Example usage:
x_value = 2.0  # Value at which to evaluate the CDF
rate_param = 1.5  # Rate parameter (lambda) of the exponential distribution

cdf_value = exponential_cdf(x_value, rate_param)
print(f"CDF at x={x_value} with rate={rate_param}: {cdf_value:.4f}")
````

10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
import math

def poisson_pmf(k, lam):
  """
  Calculates the probability mass function (PMF) of the Poisson distribution.

  Args:
    k: The number of occurrences.
    lam: The rate parameter (lambda) of the Poisson distribution.

  Returns:
    The PMF value at k.
  """
  if k < 0:
    return 0.0  # PMF is 0 for k < 0

  return (lam ** k * math.exp(-lam)) / math.factorial(k)

# Example usage:
k_value = 3  # Number of occurrences
lambda_param = 2.5  # Rate parameter (lambda)

pmf_value = poisson_pmf(k_value, lambda_param)
print(f"PMF at k={k_value} with lambda={lambda_param}: {pmf_value:.4f}")
````
